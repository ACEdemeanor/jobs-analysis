```python
#WEEK 2
#Loading dataset using pandas

import pandas as pd
df = pd.read_excel("C:/Users/AAAA/Desktop/DSA1080/jobs.csv.xlsx")
df.head()
```

```python
#Handling missing values

df.info()
df.describe()
df.columns

#cleaning the data by checking for missing values
df.isnull().sum()

df.dropna()

```

```python
#Remving duplicates
df.duplicated().sum()
df.drop_duplicates()
```

```python

#checking for the correct data types
df.dtypes


```

```python
#converting correct date format
df['posting_date'] = pd.to_datetime(df['posting_date'], errors='coerce')

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_") #---cleaning column names

```

```python
#saving the cleaned dataset to 
df.to_csv("jobs_cleaned.csv", index=False)
```

```python
#WEEK 3
#EXPLORATORY DATA ANALYSIS

#Loading the cleaned dataset
import pandas as pd
df = pd.read_csv("jobs_cleaned.csv")    


```

```python
#Summary statistics of the dataset
#this is ment to show the central tendency and the spread of the salary
df.describe()


```

```python
#grouping and aggregating the data to find the average salary by job title
average_salary_by_job = df.groupby('job_title')['salary_usd'].mean().sort_values(ascending=False)
average_salary_by_job
```

```python
#grouping and aggregating the data to find job counts by location
job_counts_by_location = df['company_location'].value_counts()
job_counts_by_location
```

```python
#VISUALIZATIONS

#1.Bar Chart---top job titles by average salary
%pip install matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
average_salary_by_job.plot(kind='bar')
plt.title('Average Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
#2.Pie Chart---distribution of job counts by location

plt.figure(figsize=(8, 8))
job_counts_by_location.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Job Counts by Location')
plt.ylabel('')
plt.show()

```

```python
import pandas as pd

# 3. Distribution PLot---distribution of salaries

df = pd.read_csv("jobs_cleaned.csv")
df['salary_usd'].plot(kind='hist', bins=30)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("visuals/salary_distribution.png")
plt.show()
```
