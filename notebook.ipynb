```python
#VISUALIZATIONS

#1.Bar Chart---top job titles by average salary
%pip install matplotlib
import matplotlib.pyplot as plt
import pandas as pd 

df=pd.read_csv("jobs_cleaned.csv")
average_salary_by_job = df.groupby('job_title')['salary_usd'].mean().sort_values(ascending=False)  

 

plt.figure(figsize=(10, 6))
average_salary_by_job.plot(kind='bar')
plt.title('Average Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("C:/Users/AAAA/Desktop/DSA1080B/visuals/average_salary_by_job.png")
plt.show()

```

```python
#2.Pie Chart---distribution of job counts by location

import matplotlib.pyplot as plt

# Get value counts of employee residence
employee_residence_counts = df['employee_residence'].value_counts()

plt.figure(figsize=(10, 8))
employee_residence_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Job Counts by Location')
plt.ylabel('')
plt.tight_layout()
plt.savefig("C:/Users/AAAA/Desktop/DSA1080B/visuals/distribution_by_location.png")
plt.show()

```

```python
# 4.Line Plot---Jobs posted over time
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("jobs_cleaned.csv")

df['posting_date'] = pd.to_datetime(df['posting_date'], errors='coerce')
jobs_over_time = df.groupby(df['posting_date'].dt.to_period('M')).size()

plt.figure(figsize=(10, 6))
jobs_over_time.plot(kind='line')    
plt.title('Jobs Posted Over Time')
plt.xlabel('Posting Date (Monthly)')
plt.ylabel('Number of Jobs')
plt.tight_layout()
plt.savefig("C:/Users/AAAA/Desktop/DSA1080B/visuals/jobs_over_time.png")
plt.show()
```

```python
# 5.Comparative Analysis---comparing average salaries between two job titles
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("jobs_cleaned.csv")


top_roles = df['job_title'].value_counts().head(5).index
sns.boxplot(data=df[df['job_title'].isin(top_roles)], x='job_title', y='salary_currency')
plt.title("Salary Comparison Across Job Roles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("C:/Users/AAAA/Desktop/DSA1080B/visuals/boxplot_salary_roles.png")
plt.show()
```

```python
#. Heatmap---correlation between numerical variables
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numerical_cols].corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("C:/Users/AAAA/Desktop/DSA1080B/visuals/correlation_heatmap.png")
plt.show()
```

```python
#WEEK 4
#ADVANCE ANALYSIS AND INSIGHT GENERATION

# Advance analysis
#1. Ranking top job titles by average salary

average_salary_by_job = df.groupby('job_title')['salary_usd'].mean().sort_values(ascending=False)
top_10_jobs = average_salary_by_job.head(10)
average_salary_by_job.head(10)


```

```python
#2. Segmentation of job market by location
job_counts_by_location = df['company_location'].value_counts()
top_locations = job_counts_by_location.head(10)
top_locations
```

```python
#3..Comparing salary trends over time for different job titles
df['posting_date'] = pd.to_datetime(df['posting_date'], errors='coerce')
salary_trends = df.groupby([df['posting_date'].dt.to_period('M'), 'job_title'])['salary_usd'].mean().unstack()  
salary_trends.head()
```

```python
#4. Demand for remote jobs over time
top_demand_roles = df['job_title'].value_counts().head(10)
top_demand_roles
```

```python
#5. Most common skills required for top job titles
top_job_titles = df['job_title'].value_counts().head(10).index  
top_job_titles
```

```python
#6. Skills and their correlation with salary
skills = df['required_skills'].str.get_dummies(sep=', ')
salary_correlation = skills.corrwith(df['salary_usd'])
salary_correlation.sort_values(ascending=False).head(10)
```

```python
#7. Experience level analysis---correlation between experience level and salary
experience_correlation = df['experience_level'].astype('category').cat.codes.corr(df['salary_usd'])
experience_correlation
```
